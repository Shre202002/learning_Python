```python
#Create New column derived from existing column

#I want to express the concentration of the station in London in mg/m3

air_quality['station_paris_in_Cubic'] = air_quality['station_paris'] *1.882
```

```python
# I want to check the ratio of the values in Paris versus Antwerp and save the result in a new column.

air_quality['ratio_antwerp_paris'] = air_quality ['station_antwerp'] / air_quality ['station_paris']
```
